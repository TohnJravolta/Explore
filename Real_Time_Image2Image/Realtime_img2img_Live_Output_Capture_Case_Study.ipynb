# Replacing the KREA canvas with another apps canvas via screen2img

Pretty straight forward but powerful. "Canvas" in this context will refer to work area etc, I will be using canvas as a placeholder for work area of any media creating tool etc. Many apps will work well simply by full screening them and then appropriately resizing that input stream inside of KREAS canvas to crop out bits. The same more or less applies to capturing a specific region of the screen to deal with uneven interfaces or other difficulties that might prevent you from just bringing in the whole window or screen. Once the selected region is an input stream in KREA via the **camera** option (by using obs virtual camera to bring in the captured screen region) the input stream can be moved around and resized to create a crop effect (hopefully the tool gets a crop object feature later).

Doing this is a good way to streamline building your canvas the way you want to get your desired output and can add or subtract whatever features for the canvas you didn't otherwise have.
